```python
#1.	Muestre en pantalla el procedimiento para importar las librerías necesarias

import pandas as pd #Leer el archivo
import numpy as np #para realizar operaciones numéricas y matemáticas
import matplotlib.pyplot as plt #Para el grafico
```

```python
#2.	Importe la data proporcionada por el docente

# Lee el archivo CSV llamado "AAPLTest2.csv" y carga sus datos en un DataFrame llamado "data"
data = pd.read_csv("AAPLTest2.csv")

# Crea una copia del DataFrame "data" y la asigna a una nueva variable llamada "data_v1"
data_v1= data.copy()
```

```python
#3.	Realiza un análisis descriptivo de la data
data_v1.describe()
```

```python
# Muestra las primeras 10 filas del DataFrame "data_v1"
data_v1.head(10)
```

```python
#4.	Convertir el campo Data de texto a datetime
data_v1.Date=pd.to_datetime(data_v1.Date, dayfirst=True) ### d-m-y
```

```python
data_v1.head()
```

```python
#5.	Asociar las fechas como índices
data_v1.set_index("Date", inplace=True)
```

```python
data_v1.head(10)
```

```python
#6.	Configurar la frecuencia de datos para convertir a una serie de tiempo
data_v1 = data_v1.asfreq("b")
data_v1.head(5)
```

```python
#7.	Encontrar el numero de valores faltantes
data_v1.isna().sum()
```

```python
#renombrar a una variable con espacio
data_v1.rename(columns={'Adj Close': 'Adj_Close'}, inplace=True)
```

```python
data_v1.isna().sum()
```

```python
#8.	Completar los valores faltantes con algún método a las variables Open, High, Low, Close y Volume
data_v1.Open = data_v1.Open.bfill()
data_v1.High = data_v1.High.ffill()
data_v1.Low = data_v1.Low.bfill()
data_v1.Close = data_v1.Close.ffill()
data_v1.Adj_Close= data_v1.Adj_Close.bfill()
data_v1.Volume = data_v1.Volume.ffill()
```

```python
data_v1.isna().sum()
```

```python
#9.	Crear una nueva columna denominada precio_cierre y copia los valores de Close
data_v1['precio_cierre'] = data_v1.Close
data_v1.head()
```

```python
#10.	Realizar el grafico de serie de tiempo de la variable precio de cierre
data_v1.precio_cierre.plot(figsize=(20,5), title="Precio Cierre")
plt.show()
```
